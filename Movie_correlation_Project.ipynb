```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# pd.options.mode.chained_assignment = None


df = pd.read_csv(r'C:\Users\valer\Downloads\movies.csv\movies.csv')
df
```

```python
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing*100)))
    
```

```python
# Check the data type.

print(df.dtypes)
```

```python
# Change data type of columns.

df['budget'].unique()
df['gross'].unique()

df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')

df.dropna(subset=['budget', 'gross'], inplace=True)

df['budget']= df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df
```

```python
df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
# Order Data a little bit to see better.

df= df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x="gross", y="budget", data=df,scatter_kws={"color":"blue"}, line_kws={"color":"black"})
```

```python
# Now check with Score and Gross.

sns.regplot(x="score", y="gross", data=df, scatter_kws={"color":"blue"}, line_kws={"color":"black"})
```

```python
# Correlation Matrix between all numeric columns

df_numeric = df.select_dtypes(include=[float, int])
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
```

```python

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
